# 1. Built-in Data Structures

## Why data structures?

Programs become powerful when they handle **collections**:
- multiple RR intervals
- many sensor readings
- a list of patients

Pythonâ€™s built-ins cover most needs:
- `list`, `tuple`
- `dict`, `set`


## Choose the right structure

- **List**: ordered, mutable (most common)
- **Tuple**: ordered, immutable (safe, hashable)
- **Dict**: key â†’ value mapping (fast lookup)
- **Set**: unique items, fast membership test

Rule of thumb: start with list/dict.


## Iteration mindset

To solve problems, you often:
1. initialize an accumulator (sum/count/max)
2. loop through items
3. update the accumulator

This pattern is everywhere.

# 3. Loop patterns (for)

## for-loop over items

`for item in collection:` is the default.

Avoid manual indexing unless you need it.


```python
rr = [0.82, 0.80, 0.79, 0.83]
for r in rr:
    print(f"RR={r:.2f}")
```

## Accumulators

Common accumulators:
- sum
- count
- max/min

We build them ourselves to learn the logic.


```python
total = 0.0
count = 0
for r in rr:
    total += r
    count += 1
mean_rr = total / count
mean_rr
```

## enumerate and range

- `range(n)` generates 0...n-1
- `enumerate(list)` gives (index, item)

Use `enumerate` for readable indexing.


```python
for i, r in enumerate(rr):
    print(i, r)
```

# 4. while loops and input loops

## when to use while

Use `while` when you donâ€™t know the number of iterations in advance.
Examples:
- keep asking until valid input
- keep looping until a condition is met


```python
i = 0
while i < 5:
    print('i =', i)
    i += 1
```

## sentinel loops

A sentinel value is a special value that means "stop".
Example: user types `q` to quit.


```python
values = []
while True:
    s = input("Enter a number (or q to quit): ")
    if s.lower() == 'q':
        break
    values.append(float(s))

print('You entered:', values)
```

# 5. Tuples, Sets, and Dictionaries

## Tuples

Tuples are immutable. Good for fixed records.


```python
point = (10, 20)
point[0], point[1]
```

## Sets

Sets store unique elements.
Fast membership check: `x in my_set`.


```python
labels = {'AF', 'Normal', 'AF', 'Noise'}
labels
```

```python
'AF' in labels, 'PVC' in labels
```

## Dictionaries

Dictionaries map keys to values. Keys must be hashable (strings, numbers, tuples).


```python
patient = {'name': 'Marlene', 'age': 65, 'hr': 78}
patient['hr']
```

```python
patient['temp'] = 36.9
patient
```
