## What is a program?

A program is a **sequence of instructions** that transforms inputs into outputs.

In Python, the most important skills early on are:
- expressing logic clearly (if/else)
- choosing good names for variables
- understanding types (int, float, str, bool)

**Rule:** If you can explain what your code does in plain language, you can usually code it.


## Variables & types

A **variable** is a name bound to a value.

Python is *dynamically typed*:
- you don‚Äôt declare types
- but types still matter at runtime

Common types:
- `int`, `float`
- `str` (text)
- `bool` (True/False)

**Key idea:** different types support different operations.


## Operators & expressions

Operators combine values into **expressions**.

- Arithmetic: `+ - * / // % **`
- Comparison: `== != < <= > >=`
- Logical: `and or not`

**Common bug:** confusing `=` (assignment) with `==` (comparison).

## Comparison operators

Comparison produces `bool` (True/False).

```python
hr = 110
hr > 100, hr == 110, hr != 80
```

## Boolean logic

- `or` requires at least one True
- `not` flips

Use parentheses to make logic clear.


```python
is_high_hr = hr > 120
is_elderly = 70 >= 65
# is_high_hr or is_elderly
not is_elderly
```

# 5. If/elif/else + nested decisions

## Control flow

`if` chooses a branch based on a condition.

Write conditions that read like English.


```python
hr = 55
if hr < 60:
    print('Low HR')
elif hr > 100:
    print('High HR')
else:
    print('Normal HR')
```

## Common pitfalls

- Missing indentation
- Using `=` instead of `==`
- Forgetting `:`

When you get an error, read the traceback.


## Nested if 

Nested if is allowed but can become hard to read.


```python
age = 70
hr = 105
if hr > 100:
    if age >= 65:
        print('High HR in older adult')
    else:
        print('High HR')
else:
    print('HR not high')
```
